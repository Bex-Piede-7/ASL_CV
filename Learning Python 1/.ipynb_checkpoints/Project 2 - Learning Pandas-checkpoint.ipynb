```python
import pandas as pd

FILE_PATH = "coffee_sales.csv"

# Import the dataset
coffee = pd.read_csv(FILE_PATH)
```

```python
# Display first 5 rows of data
print(coffee.head(5))
print()
print("----------------------------------------------------------------------------------------------")
print()

# Display last 3 rows
print(coffee.tail(3))
```

```python
# Summarize the Data
coffee.info()
```

```python
# Get basic stats
coffee.describe()
```

```python
# Display "Product" Column
print(coffee[['Product']])

print()
print("----------------------------------------------------------------------------------------------")
print()

# Display "Quantity" Column
print(coffee[['Quantity']])
```

```python
# Filter and display all sales where the product was "Latte"
print(coffee[coffee['Product'] == 'Latte'])

print()
print("----------------------------------------------------------------------------------------------")
print()

# Filter and display all sales where the quanitity was greater than 2
print(coffee[coffee['Quantity'] >= 2])

print()
print("----------------------------------------------------------------------------------------------")
print()

# Filter and display all sales where the payment method was "Cash" and the customer rating was 5
print(coffee[(coffee['Payment_Method'] == "Cash") & (coffee['Customer_Rating'] == 5)])
```

```python
# Sort the DataFrame by quantity in ascending order
print(coffee.sort_values(by='Quantity', ascending = True))

print()
print("----------------------------------------------------------------------------------------------")
print()

# Sort the DataFrame by price in descending order
print(coffee.sort_values(by='Price', ascending = False))

```

```python
# Sum Quantities
print(coffee[['Quantity']].sum())

print()
print("----------------------------------------------------------------------------------------------")
print()

# Average Customer Ratings
print(coffee[['Customer_Rating']].mean())

print()
print("----------------------------------------------------------------------------------------------")
print()

# Determine the most frequenly sold product


print()
print("----------------------------------------------------------------------------------------------")
print()

# Find the sales persons who made the most sales
```
