```python
                                                             #  INTERMEDIATE GRADE CALCULATOR - Using pandas  #
# Setup
import pandas as pd                                                            # for importing data

FILE_PATH = r"Class 1 Grades.xlsx"                                             # Don't need whole file path bc it's in the same place as the code's file

# Welcome statement
print("Welcome: You are now using the Student Grade Calculator")

# Adjust Weights
hw = .1
quiz = .2
mid = .3
final = .4

#Getting Student Info
grades_df = pd.read_excel(FILE_PATH)
grades_df['Quiz Avg'] = grades[['Quiz 1', 'Quiz 2', 'Quiz 3', 'Quiz 4']].mean(axis = 1).round(0).astype(int)
grades_df['HW Avg'] = grades[['HW 1', 'HW 2', 'HW 3', 'HW 4', 'HW 5', 'HW 6', 'HW 7', 'HW 8', 'HW 9', 'HW 10']].mean(axis = 1).round(0).astype(int)
grades_df['Final Avg'] = ((grades['HW Avg']*hw) + (grades['Quiz Avg']*quiz) + (grades['Midterm Exam']*mid)+ (grades['Final Exam']*final)).round(2)

# Constructing Letter Grade System

minA = 90
minB = 80
minC = 70
minD = 60

grade = [0, minD, minC, minB, minA, 100]
letters = ['F', 'D', 'C', 'B', 'A']

grades_df['Final Grade'] = pd.cut(
    grades_df['Final Avg'],
    bins=grade,
    labels=letters,
    right=True,
    include_lowest=True
)

grades_df = grades_df[['Student Name', 'Quiz 1', 'Quiz 2', 'Quiz 3', 'Quiz 4', 'Quiz Avg', 'HW 1', 'HW 2', 'HW 3', 'HW 4', 'HW 5', 'HW 6', 
                 'HW 7', 'HW 8', 'HW 9', 'HW 10', 'HW Avg', 'Midterm Exam', 'Final Exam', 'Final Avg', 'Final Grade']]

grades_df.head(10)

```

```python
                                                             #  BASIC GRADE CALCULATOR  #
def basicAvg(t1, t2, t3):
    sum = t1 + t2 + t3
    avg = sum / 3
    return avg

# Welcome statement
print("Welcome: You are now using the Student Grade Calculator")

# Getting Student Name
print("Input student's full name")
name = input()
print(name)

# Getting grades
print("Input student's Test 1 score")
test1 = float(input())
print("Input student's Test 2 score")
test2 = float(input())
print("Input student's Test 3 score")
test3 = float(input())

# Averaging
avg = basicAvg(test1, test2, test3)

# Assigning Letter Grade
if avg >= 90:
    letter = "A"
elif avg < 90 and avg >=80:
    letter = "B"
elif avg < 80 and avg >=70:
    letter = "C"
elif avg < 70 and avg >=65:
    letter = "D"
else: 
    letter = "F"


studentData = [
    ["Name", "Test 1", "Test 2", "Test 3", "Final Average", "Final Grade"],
    [name, test1, test2, test3, avg, letter]
]

# Print student info
print()
print()
print()
print(f"{studentData[0][0]:<20} {studentData[0][1]:<15} {studentData[0][2]:<15} {studentData[0][3]:<15} {studentData[0][4]:<15} {studentData[0][5]:<15}")
print("----------------------------------------------------------------------------------------------------")
print(f"{studentData[1][0]:<20} {studentData[1][1]:<15} {studentData[1][2]:<15} {studentData[1][3]:<15} {studentData[1][4]:<15} {studentData[1][5]:<15}")
```
