```python
                                    # Imports #
import numpy as np
import sympy
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import butter, lfilter

print("Imports Completed...")

                                    # Simulate Data #
# Array of x-values
x_data = np.linspace(0, 10, 50)

# Quadratic Eq Parameters
a = 2
b = -5
c = 3

# Generate y-values
y_true = a*x_data**2 + b*x_data + c

# Add Noise to the Data
y_data = y_true + np.random.normal(0, 10, 50)

print("Data Simulated...")
print("Complete")
```

```python
                                        # Visualize #
# Set the size
plt.figure(figsize=(8, 6))

# Create the plot
plt.scatter(x_data, y_data)
plt.title("Scatter Plot")
plt.xlabel("x-values")
plt.ylabel("y-values")

# Display the Plot
plt.show()
```
